```python
import os
import pandas as pd
from utils.ami_utils import ami, cross_ami, plot_ami, plot_cross_ami
from utils.fnn_utils import fnn, plot_fnn
```

```python
# Load data from csv file
data_x = pd.read_csv(os.path.join('data', 'mike_circ_x.txt'), sep='\t', header=None)

# Compute AMI
ami_values = ami(data_x, 0, 200)

# Plot AMI
plot_ami(ami_values)
```

```python
# Load data from csv file
data_y = pd.read_csv(os.path.join('data', 'mike_circ_y.txt'), sep='\t', header=None)

# Compute AMI
cross_ami_values = cross_ami(data_x, data_y, 0, 200)

# Plot AMI
plot_cross_ami(cross_ami_values)
```

```python
fnn_ds, fnn_percent = fnn(data_x, 20, 1, 10)

plot_fnn(fnn_ds, fnn_percent)
```
